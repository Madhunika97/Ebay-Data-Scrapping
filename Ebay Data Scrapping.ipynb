```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
try:
    source = requests.get('https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=Iphone&_sacat=0')
    
    # In order to capture any error in the URL, we can use raise_for_status function
    source.raise_for_status()
    
    soup = BeautifulSoup(source.text, 'html.parser')  # In order to fetch the HTML Text we use the .text here
    print(soup)

except Exception as e:
    print(e)
```

```python
products = soup.find(id="srp-river-main", class_="srp-river-main clearfix").find_all('li')
print(len(products))
```

```python
for product in products:
    item = product.find('span', {'role': 'heading', 'aria-level': '3'})
    price = product.find('span', class_='s-item__price')
    star_rating = product.find('div', class_='x-star-rating')
    
if item and price and star_rating:
        item_text = item.text.strip()
        price_text = price.text.strip()
        star_rating_text = star_rating.text.strip()

        print(f"Item: {item_text}, Price: {price_text}, Star Rating: {star_rating_text}")
```
